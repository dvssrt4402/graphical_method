```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def list_multiplication(l1,l2):
    l=[]
    if len(l1)==len(l2):
        for i in range(len(l1)):
            l.append(l1[i]*l2[i])
    return sum(l)


#-----multiplies corresponding values from 2 lists--------------

def less_than(a,b,c):
    if list_multiplication(a,b)<=c:
        return True

def greater_than (a,b,c):
    if list_multiplication(a,b)>=c:
        return True
    
def equal_to(a,b,c):
    if list_multiplication(a,b)==c:
        return True
    
    
#----checks the inequality constraints-----------
    
#----main code begins-----------


def graphical_method_LPP():
    
    #-------inputs------------------
    requirement= input("Enter 'max' for Maximization or 'min' for Minimization: ")
    
    
    # count of variables........
    n_obj=int(input("Enter no. of decision variables (2-variables) : "))

    # Number of constraint equations.....
    n_constraint=int(input("Enter no. of constraint equations to be taken:"))


    obj_coeff=[]
    inequality_list=[]
    l=[]
    b=[]
    d=dict()
    
    print("Enter coefficients for objective function: ")
    for i in range(n_obj):
        obj_coeff.append(int(input()))
        
        
        
    print("Enter coefficients of {} constraint equations: ".format(n_constraint))
    for i in range(n_obj):
        l=[]
        print('x{} coefficients: '.format(i+1))
        for j in range(n_constraint):
            l.append(float(input()))
        d[i+1]=l
    
    
    # Inequality List.......<=, >= and =
    
    print("Enter {} inequality constraints '<=' or '>=' or '=' for each constraint equation :".format(n_constraint))
    for i in range(n_constraint):
        inequality_list.append(input())

    

    print("Resource variables: ")
    for i in range(n_constraint):
        b.append(float(input()))
    d['rv']=b
    
    d1=dict()
    for i in range(n_obj):
        l1=[]
        for j in range(n_constraint):
            try:
                l1.append(d['rv'][j]/d[i+1][j])
            except:
                l1.append(d[i+1][j])
        d1[i+1]=l1
        
    
    x=list(d1.values())
    
    sol_list=[]
    plt.figure(figsize=[7.5,6])
    for i in range(len(x[0])):
        for j in range(len(x)):
            if j ==0:
                x1=np.array([x[j][i],0])
                sol_list.append([x[j][i],0])
            else:
                x2=np.array([0,x[j][i]])
                sol_list.append([0,x[j][i]])
        
        if x1[0]==0:  
            plt.axhline(x2[1],marker='*',Linewidth='0.75',color='red') # Plots horizontal line through a point.....
        elif x2[1]==0:
            plt.axvline(x1[0],marker='*',Linewidth='0.75',color='green') # PLots vertical line throuh a point.....
        
        else:
            plt.plot(x1,x2,marker='*',Linewidth='0.75') 
    plt.grid()
    plt.legend(['Machine C','Machine D','Machine E'])
    plt.xlabel('X1',size='20')
    plt.ylabel('X2',size='20')
    
    coeffs= list(d.values())
    
    
    
    d2=dict()
    for i in range(n_constraint):
        l2=[]
        for j in range(n_obj):
            l2.append(coeffs[j][i])
        d2[coeffs[-1][i]]=l2
    con_list=[]



    #     a=np.array([[2,1],[3,4]])
    #     b=np.array([300, 509])
    #     x= np.linalg.solve(a,b)
    
    
    for i,j in d2.items():
        con_list.append([j,i])
        
    intersection_list=[]
    for i in range(len(con_list)-1):
        try:
            for j in range(i+1,len(con_list)):
                a=np.array([con_list[i][0],con_list[j][0]])
                b=np.array([con_list[i][1],con_list[j][1]])
                intersection_list.append(list(np.linalg.solve(a,b)))
        except:
            pass

    
    
    best=[]
    coe=[]
    for i in con_list:
        coe.append(i[0])
    res=[]
    for i in con_list:
        res.append(i[1])
    sol_list+=intersection_list
    sol_list  

    
    
    
    for i in sol_list:
        c=0
        for j in range(len(coe)):
            if inequality_list[j]=='<=':
                if less_than(i,coe[j],res[j]):
                    c+=1
                else:
                    break
            elif inequality_list[j]=='>=':
                if greater_than(i,coe[j],res[j]):
                    c+=1
                else:
                    break
            elif inequality_list[j]=='=':
                if equal_to(i,coe[j],res[j]):
                    c+=1
                else:
                    break     
        if c==n_constraint:
            best.append(i)
            
    
    if len(best)==0:
        return "No Feasible Solution!"
    
    #--------------------------------------------------------------------------
    else:
        optimal_list=[]
        for i in best:
            if i[0]<0 or i[1]<0:
                continue
            else:
                optimal_list.append([((obj_coeff[0]*i[0])+(obj_coeff[1]*i[1])),i])
        if len(optimal_list)==0:
            return "No Feasible Solution!"
        if inequality_list.count('<=')==n_constraint:
            optimal_list.append([0,[0,0]])
        elif inequality_list.count('>=')==n_constraint and requirement=='max':
            return "For Maximization, the solution obtained for this problem would be an Unbounded Solution.!"
        
        #----------------------------------------------
        
        
        final=[]
        if requirement=='max':
            for i in optimal_list:
                if i[0]==max(optimal_list)[0]:
                    final.append(i)
    
        elif requirement=='min':
            for i in optimal_list:
                if i[0]==min(optimal_list)[0]:
                    final.append(i)
#     print(optimal_list)

     
#     print(final)
    if len(final)==0:
        return "No Feasible Solution!"
    else:
        for i in final:
            print("Optimal Solution is {} at point(s) {}".format(i[0],i[1]))
    

    plt.figure(figsize=[10,10])
    for i in range(len(x[0])):
        for j in range(len(x)):
            if j ==0:
                x1=np.array([x[j][i],0])
                sol_list.append([x[j][i],0])
            else:
                x2=np.array([0,x[j][i]])
                sol_list.append([0,x[j][i]])

        if x1[0]==0:
            plt.axhline(x2[1],marker='*',Linewidth='2',color='red')
        elif x2[1]==0:
            plt.axvline(x1[0],marker='*',Linewidth='2',color='green')
        else:
            plt.plot(x1,x2,marker='*',Linewidth='2')
    plt.grid()
    plt.title('Graphical Method for Optimal Solution', size='30')
    plt.legend(['Machine C','Machine D','Machine E'])
    plt.xlabel('X1',size='20')
    plt.ylabel('X2',size='20')
    for i in range(len(final)):
        plt.scatter(final[i][1][0],final[i][1][1], marker='*', color='crimson',s=300)
        plt.annotate('{} at {}'.format(final[0][0],final[i][1]),(final[i][1][0],final[i][1][1]),size='16', color='black')
    return 'Done!'
```
